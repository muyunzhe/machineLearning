```python
import pandas as pd
data_bad = pd.read_table('/home/zhongjuliu/database/金融/模型/badcase/inloan_orders_risk_predict_recent_merge_train_badcase.txt')
data_nor = pd.read_table('/home/zhongjuliu/database/金融/模型/badcase/inloan_orders_risk_predict_recent_merge_train.txt')
```

```python
print data_bad.shape
print data_nor.shape
```

```python
data_bad_des = data_bad.describe()
data_nor_des = data_nor.describe()
```

```python
data_bad_std = data_bad_des.loc[['std', 'mean']].rename(index={'std': 'std1', 'mean': 'mean1'})
data_nor_std = data_nor_des.loc[['std', 'mean']]
```

```python
describe_diff = pd.concat([data_bad_std, data_nor_std])
```

```python
std_mean_sorted = describe_diff.T.sort_values(by='std1')
```
